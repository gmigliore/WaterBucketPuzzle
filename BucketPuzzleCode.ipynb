### Water Bucket Puzzle
Suppose you have 8L, 5L and a 3L bucket with you. The buckets have no measurement lines on them, 
and you are asked to measure exactly 4L of water using the three buckets. How could you measure 
exactly 4L water using only those buckets and provided you have as much extra water as you need?

Intructions:
The game will ask you to "fill, pour, or empty." Type the full word, in lowercase for these next line.
It will then ask which bucket you would like to act on, type "8 or 5 or 3".
Once you get one of the buckets to equal four the game will stop.

```python
goal=4                                                              #The value that must be Achieved
buckets={'8':0,'5':0,'3':0}                                         #Dictionary with the current values of empty
while True:                                                         #while True to continue a loop till it is told to break
    if buckets['8'] == goal or buckets['5'] == goal:                #The Condition to break the Loop, when '8'=0 or '5'=0
        break
    else:
        action=input("Would you like to fill, pour, or empty? ")    #Question for Player on what action they would like to do.
        if action=='fill':                                          #Fill - max the bucket to the top
            which=input('Which bucket would you like to fill? ')
            if which=='8':
                buckets['8']=8
            elif which=='5':
                buckets['5']=5
            else:
                buckets['3']=3
        elif action=='pour':                                        #Move contents of bucket into another bucket
            which=input('Which bucket would you like to pour from? ')
            where=input('Which bucket would you like to pour to? ')
            if which=='8' and where=='5':
                pour=buckets['8']
                if buckets['8']+buckets['5']>5:
                    pour_into=buckets['5']
                    buckets['5']=5
                    buckets['8']=pour-(buckets['5']-pour_into)
                else:
                    buckets['8']=buckets['8']-pour
                    buckets['5']=pour+buckets['5']
            elif which=='8' and where=='3':
                pour=buckets['8']
                if buckets['8']+buckets['3']>3:
                    pour_into=buckets['3']
                    buckets['3']=3
                    buckets['8']=pour-(buckets['3']-pour_into)
                else:
                    buckets['3']=pour+buckets['3']
                    buckets['8']=buckets['8']-pour
            elif which=='5' and where=='8':
                pour=buckets['5']
                if buckets['5']+buckets['8']>8:
                    pour_into=buckets['8']
                    buckets['8']=8
                    buckets['5']=pour-(buckets['8']-pour_into)
                else:
                    buckets['8']=pour+buckets['8']
                    buckets['5']=buckets['5']-pour
            elif which=='5' and where=='3':
                pour=buckets['5']
                if buckets['5']+buckets['3']>3:
                    buckets['3']=3
                    buckets['5']=pour-buckets['3']
                else:
                    buckets['3']=pour+buckets['3']
                    buckets['5']=buckets['5']-pour
            elif which=='3' and where=='8':
                pour=buckets['3']
                if buckets['3']+buckets['8']>8:
                    pour_into=buckets['8']
                    buckets['8']=8
                    buckets['3']=pour-(buckets['8']-pour_into)
                else:
                    buckets['8']=pour+buckets['8']
                    buckets['3']=buckets['3']-pour
            elif which=='3' and where=='5':
                pour=buckets['3']
                if buckets['3']+buckets['5']>5:
                    pour_into=buckets['5']
                    buckets['5']=5
                    buckets['3']=pour-(buckets['5']-pour_into)
                else:
                    buckets['5']=pour+buckets['5']
                    buckets['3']=buckets['3']-pour
        else:
            empty=input('Which bucket would you like to empty? ')    #Empty the bucket that the player chooses
            buckets[empty]=0
        print(buckets)                                               #Prints the active bucket count after each player move
print('Congratulations, you were able to get 4 Liters into one of the Buckets!!') #Prints Congradutations.
```
